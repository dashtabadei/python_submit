```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
names = ['sepal-length', 'sepal-width', 'petal-length', 'petal-width', 'Class']
dataset = pd.read_csv(url, names=names)
print(dataset.head())
print(f"The dataset has {dataset.shape[0]} rows and {dataset.shape[1]} columns.")
le = LabelEncoder()
dataset['Class_encoded'] = le.fit_transform(dataset['Class'])
plt.figure(figsize=(8, 6))
plt.scatter(dataset['sepal-length'], dataset['sepal-width'], c=dataset['Class_encoded'])
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.colorbar()
plt.title('Iris Dataset Scatter Plot')
plt.show()
```
